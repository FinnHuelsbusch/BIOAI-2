```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
import os 
```

```python
jsons = []
# iterate through all files in the train directory
for file in os.listdir('train'):
    if file.endswith('.json'):
        with open('train/' + file) as f:
            data = json.load(f)
            jsons.append(data)
```

```python
for json in jsons:
    print(json['instance_name'])
    patients = json['patients']
    x_patient = []
    y_patient = []
    for key, value in patients.items():
        x_patient.append(value['x_coord'])
        y_patient.append(value['y_coord'])
    plt.scatter(x_patient, y_patient)
    # depot
    plt.scatter(json['depot']['x_coord'], json['depot']['y_coord'], color='red')
    plt.show()
```

```python
# check if the distance matrix is symmetric
for json in jsons:
    distance_matrix = json['travel_times']
    for i in range(len(distance_matrix)):
        for j in range(len(distance_matrix[i])):
            if distance_matrix[i][j] != distance_matrix[j][i]:
                print('Distance matrix is not symmetric for instance ' + json['instance_name'])
                
```
